```python
from types import CoroutineType
def euclid_distance(x1,y1,x2,y2):
  distance = math.sqrt((x1-x2)**2+ (y1-y2)**2)
  return float(distance)

def GetPrediction(points,label_for_points,test_points,k):
  dic = dict()
  for t in range(len(points)):
    dic[points[t]] = label_for_points[t]
  prediction = list()
  for x in test_points:
    px = x[0]
    py = x[1]
    dist_from_points = list()
    for p in points:
      dist = euclid_distance(px,py,p[0],p[1])
      temp = list()
      temp.append(dist)
      temp.append(p[0])
      temp.append(p[1])
      dist_from_points.append(temp)
    dist_from_points.sort()
    dist_form_points = dist_from_points[0:k]
    count0 =0 
    count1 =0 
    count2 =0
    for i in dist_from_points:
      if dic[dist_from_points[1:]] == 0:
        count0 = count0 + 1
      elif dic[dist_from_points[1:]] == 1:
        count1 = count1 + 1
      else:
        count2 = count2 + 1
    if count0 > count1 and count0 > count2:
      prediction.append(0)
    elif count1 > count0 and count1 > count2:
      prediction.append(1)
    else:
      prediction.append(2)
  return prediction
```
